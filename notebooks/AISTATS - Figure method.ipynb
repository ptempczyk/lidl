```python
import numpy as np
from scipy.special import erf
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import colors

def p(x, b, a, s):
    return 1 / (2 * (b - a)) * (erf((b - x) / np.sqrt(2 * s**2)) - erf((a - x) / np.sqrt(2 * s**2)))

def gauss(x, s):
    return 1/np.sqrt(2*np.pi*s**2)*np.exp(-1/2*(x**2/s**2))

def dens_square(x_1, x_2, b, a, s):
    return p(x_1, b,a,s) * p(x_2, b,a,s)

def dens_line(x_1, x_2, b, a, s):
    return p(x_1, b,a,s) * gauss(x_2, s)

def dens_point(x_1, x_2, b, a, s):
    return gauss(x_1, s) * gauss(x_2, s)

matplotlib.rc('text', usetex=True)
plt.rcParams['font.family'] = 'serif'
label_size = 16
ticks_size = 16
legend_size = 16

cmap = "magma"

b, a = 1, -1

x = np.linspace(a-1, b+1, 201)
y = np.linspace(a-1, b+1, 201)
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(15,12))

s = 1e-5

Z = dens_square(X, Y, b, a, s)
ax = plt.subplot(3,3,1)
ax.axes.get_yaxis().set_visible(True)
ax.axes.get_xaxis().set_visible(False)
cmap_custom = colors.ListedColormap(['black', '#fbfabd'])
plt.contourf(X, Y, Z, 20, cmap=cmap_custom)
plt.colorbar(ticks=[])
plt.tick_params(labelsize=ticks_size)

Z = dens_line(X, Y, b, a, s)
plt.subplot(3,3,2)
plt.axis("off")
cmap_custom = colors.ListedColormap(['black', '#fbfabd'])
plt.contourf(X, Y, Z, 20, cmap=cmap_custom)
plt.colorbar(ticks=[])
plt.tick_params(labelsize=ticks_size)

Z = dens_point(X, Y, b, a, s)
plt.subplot(3,3,3)
plt.axis("off")
cmap_custom = colors.ListedColormap(['black', '#fbfabd'])
plt.contourf(X, Y, Z, 20, cmap=cmap_custom)
plt.colorbar(ticks=[])
plt.tick_params(labelsize=ticks_size)


s = 0.1
Z = dens_square(X, Y, b, a, s)
ax = plt.subplot(3,3,4)
ax.axes.get_yaxis().set_visible(True)
ax.axes.get_xaxis().set_visible(False)
plt.contourf(X, Y, Z, 20, cmap=cmap)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 2)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
sqmax = np.max(Z)
plt.clim(0, sqmax)
ax.set_ylabel("$x_2$", size=label_size)

Z = dens_line(X, Y, b, a, s)
plt.subplot(3,3,5)
plt.axis("off")
plt.contourf(X, Y, Z, 20, cmap=cmap)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 0)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
limax = np.max(Z)
plt.clim(0, limax)

Z = dens_point(X, Y, b, a, s)
plt.subplot(3,3,6)
plt.axis("off")
plt.contourf(X, Y, Z, 20, cmap=cmap)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 0)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
pomax = np.max(Z)
plt.clim(0, pomax)

s = 0.2
Z = dens_square(X, Y, b, a, s)
ax = plt.subplot(3,3,7)
plt.contourf(X, Y, Z, 20, cmap=cmap)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 2)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
plt.clim(0, sqmax)
ax.axes.get_yaxis().set_visible(True)
ax.axes.get_xaxis().set_visible(True)

Z = dens_line(X, Y, b, a, s)
ax = plt.subplot(3,3,8)
plt.contourf(X, Y, Z, 20, cmap=cmap)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 0)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
plt.clim(0, limax)
ax.axes.get_yaxis().set_visible(False)
ax.axes.get_xaxis().set_visible(True)
ax.set_xlabel("$x_1$", size=label_size)


Z = dens_point(X, Y, b, a, s)
ax = plt.subplot(3,3,9)
plt.contourf(X, Y, Z, 20, cmap=cmap)
plt.clim(0, pomax)
cbar = plt.colorbar(ticks=[0, np.around(np.max(Z), 0)])
cbar.ax.tick_params(labelsize=ticks_size)
plt.tick_params(labelsize=ticks_size)
ax.axes.get_yaxis().set_visible(False)
ax.axes.get_xaxis().set_visible(True)
```
