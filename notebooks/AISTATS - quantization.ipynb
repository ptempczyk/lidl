```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,1.5))
x = np.random.rand(500)*10
y = 10 * np.sin(x/3.15)
plt.subplot(1,3,1)
plt.scatter(x, y, s=5, c='darkorange')
xr, yr = np.round(x), np.round(y)
plt.xticks([])
plt.yticks([])
plt.subplot(1,3,2)
plt.scatter(xr, yr, s=5, c='darkorange')
plt.xticks([])
plt.yticks([])
plt.subplot(1,3,3)
xrn, yrn = xr + np.random.rand(*xr.shape), yr + np.random.rand(*yr.shape)
plt.scatter(xrn, yrn, s=5, c='darkorange')
plt.xticks([])
plt.yticks([])
plt.show()
```
